```python

import mysql.connector
import pandas as pd

# -------------------------------------------
# BLOQUE 1: CONEXI√ìN A LA BASE DE DATOS
# -------------------------------------------
try:
    conexion = mysql.connector.connect(
    host='localhost',
    port=3306,
    user='root',
    password='clave',
    database='patata'
)
        
    if conexion.is_connected():
        print("\n¬°Conexi√≥n exitosa! üöÄ")
except mysql.connector.Error as e:
    print(f"\n‚ùå Error de conexi√≥n: {e}")
    # Salimos del script si no hay conexi√≥n
    exit(1)


# -------------------------------------------
# BLOQUE 2: EJEMPLO DE OTRA OPERACI√ìN SQL(ELIMINACION DE BASE DE DATOS)
# -------------------------------------------
try:
    # -------------------------------------------
    # EJEMPLO DE OTRA OPERACI√ìN (ELIMINACION DE BASE DE DATOS):
    cursor = conexion.cursor()
    
    # Verificar primero si la base de datos existe
    cursor.execute("SHOW DATABASES LIKE 'patata'")
    resultado = cursor.fetchone()
    
    if not resultado:
        # Lanzar manualmente una excepci√≥n si no existe
        raise Exception("La base de datos 'patata' no existe")
    
    # Si existe, proceder a eliminarla
    cursor.execute("DROP DATABASE patata")
    print("Database eliminada")
    cursor.close()
    # -------------------------------------------
    
except Exception as e:
    # Capturar el error personalizado de base de datos inexistente
    if "La base de datos 'patata' no existe" in str(e):
        print(f"\n‚ùå Error: {e}")
    else:
        print(f"\n‚ùå Error durante operaciones SQL: {e}")

# -------------------------------------------
# BLOQUE 3: CIERRE DE CONEXI√ìN
# -------------------------------------------
if conexion.is_connected():
    conexion.close()
    print("\nConexi√≥n cerrada correctamente.")
```
